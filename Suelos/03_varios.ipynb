```python
Columnas = list(Frame)
Col2 = ["Arenas", "Arcillas", "Limos"]
```

```python
aux = []
for i in range(3):
    aux0 = [     Columnas[6+i], "Total",  Frame.iloc[:,6+i].mean(),   Frame.iloc[:,6+i].std()]
    aux.append(aux0)
    aux0 = [Columnas[6+i],"Norte", Frame_N.iloc[:,6+i].mean(), Frame_S.iloc[:,6+i].std()]
    aux.append(aux0)
    aux0 = [Columnas[6+i], "Sur" , Frame_S.iloc[:,6+i].mean(), Frame_S.iloc[:,6+i].std()]
    aux.append(aux0)

frame_export = pd.DataFrame(aux)
frame_export.columns = ["Col", "Zone", "Mean", "STD"]

frame_export.to_excel(writer, "porcentajes", index=False)
```
